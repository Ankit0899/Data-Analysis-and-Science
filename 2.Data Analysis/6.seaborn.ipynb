```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
## Create a Scatter Plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs tip")

```

## Line Plot

```python
sns.lineplot(x='size',y='total_bill', data=tips)
plt.title("Line plot of Total Bills By Size")
```

```python
## Categorical Plots
## Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plat of total bill ')
plt.show
```

## Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

### Histograms

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

### PairPlot

```python
sns.pairplot(tips)
```

### HeatMap

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('data.csv')
sales_df.head()
```

```python
## Plot total sales by Product

plt.figure(figsize=(10,6))
sns.barplot(x='Category',y="Value",data=sales_df,estimator=sum)
plt.title('Total Sales By Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Value",data=sales_df,estimator=sum)
plt.title('Total Sales By Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
