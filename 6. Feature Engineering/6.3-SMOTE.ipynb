### SMOTE(Synthetic Minority Oversampling Technique)

Smote is a technique used in machine learning to address inbalanced dataset where the minority class significantly fewer instances than the majority class .
SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
import sklearn
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                    weights=[0.90],random_state=12) 
```

```python
import pandas as pd

df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])

final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```
